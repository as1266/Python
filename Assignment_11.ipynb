```
import pymysql
import random

db = pymysql.connect("localhost", "root", "Computer34_", "pydb")
cursor = db.cursor()


def addStudents():
    studentEntries = int(input("How many students would you like to add? Press 0 if none: "))
    i = 0
    if studentEntries > 0:
        while i < studentEntries:
            name = input("Please enter a student name into the database: ")
            major = input("Please enter the student's major into the database: ")
            sql = "insert into tstudent(stuname,major,stuyear,tuition) values(%s,%s,%s,%s)"
            values = (name, major, "2020-04-06", 2100.0)
            print(values)
            cursor.execute(sql, values)
            db.commit()
            i = i + 1
    else:
        return
    cursor.close()
    db.close()
    return


def deleteStudents():
    studentId = int(input("Please enter a student ID to delete them from the database, Press 0 if none: "))
    if studentId > 0:
        sql = "Delete From tstudent Where tstudentid = %s "
        value = studentId
        print(value)
        cursor.execute(sql, value)
        db.commit()
        cursor.close()
        db.close()
        return
    else:
        return


def modify():
    studentId = int(input("Please enter a student ID to update them in the database,Press 0 if none: "))
    if studentId > 0:
        studentName = input("Please enter the new name for the student chosen")
        sql = "Update tstudent set stuname=%s Where tstudentid = %s "
        value = (studentName, studentId)
        print(value)
        cursor.execute(sql, value)
        db.commit()
        cursor.close()
        db.close()
        return
    else:
        return


def majorCount():
    answer = str(input("Do you want the total count of a specific major? Press Y or N: "))
    if answer == "y" or answer == "Y":
        studentMajor = str(input("Please enter the major you want the count of"))
        sql = "Select major , count(*) FROM tstudent where major=%s "
        value = (studentMajor)
        cursor.execute(sql, value)
        data = cursor.fetchall()
        print("Major          Count")
        print("------------------------")
        for i in data:
            print(i[0], i[1])
        cursor.close()
        print("------------------------")
        return
    else:
        return


def randomStudent():
    answer = str(input("Do you want to get a random students name and id?Press Y or N: "))
    if answer == "y" or answer == "Y":
        sql = "Select stuname, tstudentid FROM tstudent WHERE tstudentid =%s"
        value = random.randint(1, 13)
        cursor.execute(sql, value)
        data = cursor.fetchall()
        print("Name      ID")
        print("------------------------")
        for i in data:
            print(str(i[0]), int(i[1]))
        print("------------------------")
        return
    else:
        return


addStudents()
deleteStudents()
modify()
majorCount()
randomStudent()
```
